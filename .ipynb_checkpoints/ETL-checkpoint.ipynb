```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import geopy.distance
import mpu

%matplotlib inline
```

```python
train = '../project4/asset/train.csv'
test = '../project4/asset/test.csv'
spray = '../project4/asset/spray.csv'
weather = '../project4/asset/weather.csv'
```

```python
df_train = pd.read_csv(train)
df_test = pd.read_csv(test)
df_weather = pd.read_csv(weather)
```

```python
df = df_train.merge(df_test, how='outer')
```

```python
df = df[['Id','Date', 'Address', 'Species', 'Block', 'Street', 'Trap',
       'AddressNumberAndStreet', 'Latitude', 'Longitude', 'AddressAccuracy',
       'NumMosquitos', 'WnvPresent']]
```

```python
df.head()
```

```python
lat1 = 41.995
lon1 = -87.933

lat2 = 41.786
lon2 = -87.752
dist = []

for i in range(len(df)):
    dist1 = mpu.haversine_distance((df['Latitude'][i], df['Longitude'][i]), (lat1,lon1))
    dist2 = mpu.haversine_distance((df['Latitude'][i], df['Longitude'][i]), (lat2,lon2))
    if dist1> dist2:
        dist.append(2)
    else:
        dist.append(1)
    
```

```python
df['Station'] = dist
```

```python
df_weather = pd.read_csv(weather)
df_weather.head()
```

```python
# Forward filling the empty rows for station 2 
df_weather['Sunrise'].replace('-',np.nan,inplace=True)
df_weather['Sunrise'].fillna(method='ffill',inplace=True)
df_weather['Sunrise'] = df_weather['Sunrise'].astype(int)

df_weather['Sunset'].replace('-',np.nan,inplace=True)
df_weather['Sunset'].fillna(method='ffill',inplace=True)
df_weather['Sunset'] = df_weather['Sunset'].astype(int)

df_weather['Depart'].replace('M',np.nan,inplace=True)
df_weather['Depart'].fillna(method='ffill',inplace=True)
df_weather['Depart'] = df_weather['Depart'].astype(float)

df_weather['SnowFall'].replace('M',np.nan,inplace=True)
df_weather['SnowFall'].replace('T',0,inplace=True)
df_weather['SnowFall'].replace('  T',0,inplace=True)
df_weather['SnowFall'].fillna(method='ffill',inplace=True)
df_weather['SnowFall'] = df_weather['SnowFall'].astype(float)

df_weather['Heat'].replace('M',np.nan,inplace=True)
df_weather['Heat'].fillna(method='ffill',inplace=True)
df_weather['Heat'] = df_weather['Heat'].astype(int)

df_weather['Cool'].replace('M',np.nan,inplace=True)
df_weather['Cool'].fillna(method='ffill',inplace=True)
df_weather['Cool'] = df_weather['Cool'].astype(int)

df_weather['PrecipTotal'].replace('M',np.nan,inplace=True)
df_weather['PrecipTotal'].replace('T',0,inplace=True)
df_weather['PrecipTotal'].replace('  T',0,inplace=True)
df_weather['PrecipTotal'].fillna(method='ffill',inplace=True)
df_weather['PrecipTotal'] = df_weather['PrecipTotal'].astype(float)

df_weather['StnPressure'].replace('M',np.nan,inplace=True)
df_weather['StnPressure'].fillna(method='ffill',inplace=True)
df_weather['StnPressure'] = df_weather['StnPressure'].astype(float)

df_weather['SeaLevel'].replace('M',np.nan,inplace=True)
df_weather['SeaLevel'].fillna(method='ffill',inplace=True)
df_weather['SeaLevel'] = df_weather['SeaLevel'].astype(float)

df_weather['AvgSpeed'].replace('M',np.nan,inplace=True)
df_weather['AvgSpeed'].fillna(method='ffill',inplace=True)
df_weather['AvgSpeed'] = df_weather['AvgSpeed'].astype(float)
```

```python
# Replace incorrect Tavg/null rows
df_weather['Tavg'] = (df_weather['Tmax']+df_weather['Tmin'])/2
```

```python
df_weather['WetBulb'].replace('M',63,inplace=True)
df_weather['WetBulb'] = df_weather['WetBulb'].astype(int)
```

```python
df_weather.groupby('Date').mean()
```

```python
df_weather['CodeSum'] = df_weather['CodeSum'].apply(lambda x: x.replace('+',''))
```

```python
has_water = ['FC','TS','GR','RA','DZ','SN','SG','GS','PL','IC','FG','BR','SH']
```

```python
lst = []
for rows in df_weather['CodeSum']:
    if any(word in rows for word in has_water):
        lst.append(1)
    else:
        lst.append(0)
df_weather['Code'] = lst
```

```python
merged = pd.merge(df,df_weather,on=('Station','Date'))
```

```python
merged.to_csv('cleaned_traintest.csv')
```
